Convert categorical variables into numerical representations using techniques like one-hot encoding.

```python
df.dtypes
```

```python
# Select categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns
print(categorical_columns)
```

```python
unique_values_summary = {}
for col in categorical_columns:
    unique_values_summary[col] = (df[col].unique(),df[col].nunique())
```

```python
unique_values_summary
```

```python
categorical_columns=[]
for i in unique_values_summary:
    if (unique_values_summary[i][1]<30):
        categorical_columns.append(i)
```

```python
categorical_columns
```

```python
df_encoded = pd.get_dummies(df, columns=categorical_columns,dtype='int')
```
