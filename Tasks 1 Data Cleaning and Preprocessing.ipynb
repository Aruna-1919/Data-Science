Handle missing values in the dataset.

```python
df.isnull().sum()
```

Convert categorical variables into numerical representations using techniques like one-hot encoding.

```python
# finding categorical columns
df.dtypes
```

```python
categorical_columns=df.select_dtypes(include=['object']).columns
```

```python
#printing categorical columns
categorical_columns
```

```python
#to know how many unique categories are there for every column
unique_count={}
for i in categorical_columns:
    unique_count[i]=df[i].nunique()
unique_count
```

```python
#consider only categorical columns whose unique values are less than 50
final_categorical_columns=[]
for col in unique_count:
    if unique_count[col] < 50 :
        final_categorical_columns.append(col)
final_categorical_columns
```

```python
#converting categorical column to numerical using one-hot encoding 
df_encode=pd.get_dummies(df,columns=final_categorical_columns)
df_encode
```

```python
print("Before applying one-hot encoding column count : ",df.shape[1])
print("After applying one-hot encoding column count : ",df_encode.shape[1])
```

```python
file_path = 'updated_dataset_after_task_1.csv'

# Save DataFrame to CSV
df_encode.to_csv(file_path, index=False)  # Set index=False to exclude row numbers from the CSV

print(f"DataFrame saved to '{file_path}'")
```
